```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('bitcoin.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(['Date'],1,inplace=True)
```

```python
data.head()
```

```python
days = 30
data['Prediction'] = data[['Price']].shift(-days)
```

```python
data.head()
```

```python
data.tail()
```

```python
X = np.array(data.drop(['Prediction'],1))
```

```python
X = X[: len(data) - days]
```

```python
X.shape
```

```python
Y = np.array(data['Prediction'])
```

```python
Y = Y[: -days]
```

```python
Y.shape
```

```python
# splitting the data into train set and test set
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2)
```

```python
p_days_array = np.array(data.drop(['Prediction'],1))[days:]
```

```python
print(p_days_array)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf = RandomForestRegressor(n_estimators=1000, random_state=1)
```

```python
Rf.fit(X_train,Ytrain)
print("Random Forest Accuracy : {:.2f}%".format(Rf.score(Xtest,Ytest)*100))
```

```python
Rf_prediction = Rf.predict(Xtest)
print(Rf_prediction)
```
